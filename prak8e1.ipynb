```python
import pandas as pd

data_rafael = pd.read_csv('C:/probstat/houseprice.csv')
print(data_rafael.head(5))
```

```python
col_list=['Price']
numhouse = data_rafael[data_rafael.columns[data_rafael.columns.isin(col_list)]] 
plt.figure(figsize=(10,5))
numhouse.boxplot(sym='r*', grid=False) 
plt.show()
```
